```python
# 유니버셜 함수 : 배열의 각원소를 빠르게 처리하는 함수
import numpy as np

arr = np.arange(10)
print (arr)

# 제곱근 구하기
np.sqrt(arr)
# 지수값 구하기 
np.exp(arr)

# 여기서 seed는 랜덤 숫자를 항상 같은 숫자가 나오도록 시작점을 잡아주는 기능을 함
rng = np.random.default_rng(seed=12345)  
data1 = rng.standard_normal(8)
data2 = rng.standard_normal(8)
print (data1)
print (data2)

# 두배열을 비교해 큰값을 return
np.maximum(data1, data2)

arr = rng.standard_normal(7)*5
print (arr)

# 분수를 받아 몫과 나머지를 반환
remainder, whole_part = np.modf(arr)
remainder
whole_part

arr
out = np.zeros_like(arr)
print (out)
a = np.add(arr, 1)
b = np.add(arr, 1, out=out)
print (f"a = {a}")
print (f"b = {b}")
out
```
